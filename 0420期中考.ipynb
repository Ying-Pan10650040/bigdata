```
# 請用SportsAnalytics套件所提供的NBA數據進行分析(任選一年)

install.packages("SportsAnalytics")  
library(SportsAnalytics)
```

```
NBA1920 <- fetch_NBAPlayerStatistics("19-20")
```

```
# 利用order()函數,選一種類型的資料出來,由大到小排序(decreasing = T) Exp:GamesPlayed

NBA1920_order <- NBA1920[order(NBA1920$GamesPlayed, decreasing = T),]
NBA1920_order
```

```
# 用subset()做欄位條件搜尋

subset(NBA1920,Team=="MIA")
```

```
# 用grepl()做字串條件搜尋

NBA1920[grepl("Allen",NBA1920$Name),]
```

```
# 用ggplot畫出,四種不同類型的圖表(Exp:散佈圖、長條圖、折線圖...)

install.packages("ggplot2")
library(ggplot2)
```

```
install.packages("devtools")
devtools::install_github("cttobin/ggthemr")
library(ggthemr)
```

```
 NBA1920 %>% 
  count(Position) %>%
    ggplot(aes(x = fct_reorder(Position, n), y = n, fill = Position)) +
    geom_col() +
    geom_col(fill = c("#80ffec","#27c7d8","#00598d","#008fb9","#1d2657")) + 
    geom_text(aes(label = n), vjust = -0.25) +
    labs(title = "不同位置之球員人數對比", x = "位置", y = "人數") + 
    theme(legend.position = "none")
```

```
ggthemr("fresh")

NBA1920 %>%
  ggplot(aes(x = TotalPoints, y = Position, fill = Position)) + 
  geom_density_ridges()
```

```
ggthemr("dust")

NBA1920 %>% 
  mutate(Position = fct_reorder(Position, TotalPoints, median, na.rm = T)) %>%
  ggplot(aes(Position, TotalPoints)) + 
  geom_point() +
  geom_boxplot() + 
  coord_flip()
```
