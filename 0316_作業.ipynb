```
## 重大職災統計

install.packages("jsonlite") # 安裝讀取json套件
library("jsonlite")  # 載入json套件
```

```
data01 <- jsonlite::fromJSON("A17000000J-030126-KxC.json")  # 讀取json檔案
str(data01)
```

```
head(data01,n=10)  # 預覽前10筆資料
tail(data01,n=10)  # 預覽後10筆資料
```

```
data01[grepl("星期五",data01$`星期`),]  # 搜尋"星期五"
data01[grepl("30-34歲",data01$`年齡`),]  # 搜尋"30-34歲"
data01[grepl("1個月以下",data01$`工作經歷`),]  # 搜尋"1個月以下"
```

```
table(data01$`時段`)  # 計算 時段
table(data01$`年度`)  # 計算 年度
table(data01$`月份`)  # 計算 月份
```
